```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import fetch_california_housing
from sklearn.cluster import KMeans
```

```python
dataset = fetch_california_housing()
X = dataset.data
df = pd.DataFrame(X, columns=dataset.feature_names)
X_selected = df[['MedInc', 'HouseAge', 'AveRooms']]
```

```python
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
clusters = kmeans.fit_predict(X_selected)
df['cluster'] = clusters

plt.figure()
scatter = plt.scatter(df['MedInc'], df['AveRooms'], c=df['cluster'], cmap='viridis', alpha=0.6)

plt.xlabel('(MedInc)')
plt.ylabel('(AveRooms)')
plt.title('KMeans Clustering of California Housing Dataset (MedInc vs AveRooms)')
plt.colorbar(scatter, label='Cluster')
plt.grid(True)
plt.show()
```
